```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(20, 15))
sb.heatmap(numeric_df.corr(method='pearson'), annot=True)
plt.show()
```

```python
numeric_df = one_year.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(15, 10))
sb.heatmap(numeric_df.corr(method='pearson'), annot=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))


sns.heatmap(numeric_df.corr(method='kendall'), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Kendall Correlation')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(method='spearman'), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation')
plt.show()
```
